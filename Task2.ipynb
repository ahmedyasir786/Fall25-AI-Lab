```python
class FizzBuzzGame:
    def __init__(self, limit):
        self.limit = limit

    def play(self):
        for i in range(1, self.limit + 1):
            output = ""
            if i % 3 == 0:
                output += "Fizz"
            if i % 5 == 0: 
                output += "Buzz"
            if output == "":
                print(i)
            else:
                print(output)


print("Fizz Buzz up to 15:")
game1 = FizzBuzzGame(15)
game1.play()

print("-" * 20)

print("Fizz Buzz up to 30:")
game2 = FizzBuzzGame(30)
game2.play()
```
